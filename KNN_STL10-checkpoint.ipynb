```python
from PIL import Image
import pandas as pd
import os
import cv2
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report

def load_images_with_labels(folder):
   
    X = []
    y = []
    
    for file_name in os.listdir(folder):
        file_path = os.path.join(folder, file_name)
        
        # قراءة الصورة
        img = Image.open(file_path).convert('RGB')
        img = img.resize((64, 64))  # تغيير الأبعاد
        img_array = np.array(img)
        X.append(img_array)
        
        # استخراج التصنيف من الرقم في نهاية اسم الملف
        try:
            label = int(file_name.split("_")[-1])  # الرقم بعد "png_"
        except ValueError:
            label = -1  # تعيين -1 إذا فشل في استخراج الرقم
        
        y.append(label)
    
    return np.array(X), np.array(y)

# المجلدات
train_folder = "C:/Users/Mariam Yasser/Desktop/STL10_Project/train_images"
test_folder = "C:/Users/Mariam Yasser/Desktop/STL10_Project/test_images"

# تحميل الصور والتصنيفات
X_train, y_train = load_images_with_labels(train_folder)
X_test, y_test = load_images_with_labels(test_folder)

print("Train data shape:", X_train.shape, "Train labels shape:", y_train.shape)
print("Test data shape:", X_test.shape, "Test labels shape:", y_test.shape)
```

```python
# تسطيح الصور (تحويلها إلى شكل 2D)
X_train = X_train.reshape(X_train.shape[0], -1)  # تسطيح X_train
X_test = X_test.reshape(X_test.shape[0], -1)    # تسطيح X_test

# تطبيع البيانات (تحويل القيم إلى [0, 1])
X_train_normalized = X_train / 255.0
X_test_normalized = X_test / 255.0

# تأكيد الأبعاد الجديدة
print("Shape of X_train after flattening:", X_train_normalized.shape)
print("Shape of X_test after flattening:", X_test_normalized.shape)

```

```python
# 1. تدريب نموذج KNN
knn = KNeighborsClassifier(n_neighbors=3)  # يمكنك تغيير قيمة n_neighbors
knn.fit(X_train_normalized, y_train)

# 2. التنبؤ بالتصنيفات
y_pred = knn.predict(X_test_normalized)

# 3. حساب الدقة (Accuracy)
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy * 100:.2f}%")
```

```python
# 1. تدريب نموذج KNN
knn = KNeighborsClassifier(n_neighbors=5)  # يمكنك تغيير قيمة n_neighbors
knn.fit(X_train_normalized, y_train)

# 2. التنبؤ بالتصنيفات
y_pred = knn.predict(X_test_normalized)

# 3. حساب الدقة (Accuracy)
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy * 100:.2f}%")
```
