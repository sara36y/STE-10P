```python
import os
import matplotlib.pyplot as plt
from PIL import Image

# المسار الرئيسي
data_path = r'C:\Users\Mariam Yasser\Desktop\STL10_Project'

# استعراض المجلدات الفرعية
subfolders = ['train_images', 'test_images', 'unlabeled_images']

# قائمة لتخزين الصور
image_files = []

# إضافة الصور من المجلدات الفرعية
for folder in subfolders:
    folder_path = os.path.join(data_path, folder)
    
    # تصفية الملفات التي هي صور فقط (تأكد من امتداد الصور في الداتاست)
    image_files.extend([os.path.join(folder_path, file) for file in os.listdir(folder_path) if file.endswith(('.png', '.jpg', '.jpeg'))])

# طباعة عدد الصور
print(f'Number of images: {len(image_files)}')

# عرض بعض الصور
if len(image_files) > 0:
    for i in range(min(2, len(image_files))):  # عرض أول 5 صور أو أقل إذا كانت القائمة صغيرة
        img_path = image_files[i]
        img = Image.open(img_path)
        plt.imshow(img)
        plt.show()
else:
    print("No images found.")

```

```python
import numpy as np

# تحويل كل صورة إلى مصفوفة أحادية البُعد
def preprocess_image(image_path):
    img = Image.open(image_path)
    img = img.resize((32, 32))  # تغيير الحجم إلى 32x32
    img = np.array(img) / 255.0  # تطبيع القيم إلى [0, 1]
    return img.flatten()  # تحويل الصورة إلى مصفوفة أحادية البُعد

# تحويل جميع الصور إلى مصفوفات أحادية البُعد
image_vectors = [preprocess_image(img_path) for img_path in image_files[:1000]]  
```

```python
from sklearn.cluster import KMeans

# تطبيق K-Means
kmeans = KMeans(n_clusters=10, random_state=42)
kmeans.fit(image_vectors)  # تدريب النموذج على البيانات

# الحصول على التصنيفات (المجموعات) لكل صورة
labels = kmeans.labels_

# عرض التصنيفات لأولى بعض الصور
print(labels[:10])  # عرض أول 10 تصنيفات
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# رسم توزيع المجموعات
sns.countplot(x=labels)
plt.title('Cluster Distribution')
plt.show()
```

```python
# تجربة عدة قيم للمجموعات (clusters) واختيار العدد الأمثل
inertia = []
for n_clusters in range(1, 11):
    kmeans = KMeans(n_clusters=n_clusters, random_state=42)
    kmeans.fit(image_vectors)
    inertia.append(kmeans.inertia_)

# رسم مخطط الكوع
plt.plot(range(1, 11), inertia, marker='o')
plt.title('Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia')
plt.show()
```

```python
from sklearn.metrics import silhouette_score

# حساب الـ Silhouette Score
score = silhouette_score(image_vectors, labels)
print(f'Silhouette Score: {score}')
```

```python
# طباعة مراكز الفئات (centroids)
print("Cluster centers (centroids):")
print(kmeans.cluster_centers_)

```

```python
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np
import os

# تحديد عدد الفئات K (مثال: 10 فئات)
k = 10

# تطبيق K-Means
kmeans = KMeans(n_clusters=k, random_state=0)

# تدريب K-Means على البيانات (صور تم تحويلها إلى مصفوفات أحادية البُعد)
kmeans.fit(image_vectors)

# الحصول على التصنيفات (labels) الخاصة بكل صورة
labels = kmeans.labels_

# تقسيم الصور إلى مجموعات بناءً على الفئات (labels)
grouped_images = {i: [] for i in range(k)}  # إنشاء قاموس لتخزين الصور لكل فئة

# تعبئة المجموعات بالصور التي تنتمي لكل فئة
for i, label in enumerate(labels):
    grouped_images[label].append(image_files[i])  # إضافة مسار الصورة إلى المجموعة المناسبة

# عرض الصور في كل مجموعة
for label, image_paths in grouped_images.items():
    print(f"Group {label + 1}:")
    
    # تحديد عدد الصور التي سيتم عرضها في هذه المجموعة (مثال: عرض أول 5 صور في كل مجموعة)
    for i, image_path in enumerate(image_paths[:5]):
        img = Image.open(image_path)
        img = img.resize((32, 32))  # لتغيير حجم الصور إذا كانت كبيرة
        plt.subplot(1, 5, i + 1)
        plt.imshow(img)
        plt.axis('off')  # إخفاء المحاور لتوضيح الصورة بشكل أفضل

    plt.show()

```
